# Cell 1 – Import Libraries

```python
import sqlite3
import pandas as pd
```

# Cell 2 – Connect to Database

```python
conn = sqlite3.connect("../db/nifty100.db")

print("Connected Successfully")
```

# Cell 3 – Show All Tables

```python
query = """
SELECT name
FROM sqlite_master
WHERE type='table'
ORDER BY name;
"""

pd.read_sql(query, conn)
```

# Cell 4 – Total Companies

```python
query = """
SELECT COUNT(*) AS Total_Companies
FROM companies;
"""

pd.read_sql(query, conn)
```

# Cell 5 – Row Counts

```python
tables = [
    "companies",
    "profitandloss",
    "balancesheet",
    "cashflow",
    "stock_prices"
]

for table in tables:
    query = f"SELECT COUNT(*) AS Rows FROM {table};"
    df = pd.read_sql(query, conn)
    print(f"{table:20} {df.iloc[0,0]}")
```

# Cell 6 – Top 10 Companies

```python
query = """
SELECT id,
       company_name,
       ticker
FROM companies
LIMIT 10;
"""

pd.read_sql(query, conn)
```

# Cell 7 – Profit & Loss Sample

```python
query = """
SELECT *
FROM profitandloss
LIMIT 10;
"""

pd.read_sql(query, conn)
```

# Cell 8 – Balance Sheet Sample

```python
query = """
SELECT *
FROM balancesheet
LIMIT 10;
"""

pd.read_sql(query, conn)
```

# Cell 9 – Cash Flow Sample

```python
query = """
SELECT *
FROM cashflow
LIMIT 10;
"""

pd.read_sql(query, conn)
```

# Cell 10 – Financial Ratios Sample

```python
query = """
SELECT *
FROM financial_ratios
LIMIT 10;
"""

pd.read_sql(query, conn)
```

# Cell 11 – Companies by Industry

```python
query = """
SELECT
industry,
COUNT(*) AS Total
FROM companies
GROUP BY industry
ORDER BY Total DESC;
"""

pd.read_sql(query, conn)
```

# Cell 12 – Latest Financial Year

```python
query = """
SELECT MAX(year) AS Latest_Year
FROM profitandloss;
"""

pd.read_sql(query, conn)
```

# Cell 13 – Foreign Key Check

```python
query = """
PRAGMA foreign_key_check;
"""

pd.read_sql(query, conn)
```

# Cell 14 – Database Summary

```python
summary = {
    "Table": [],
    "Rows": []
}

tables = [
    "companies",
    "profitandloss",
    "balancesheet",
    "cashflow",
    "analysis",
    "documents",
    "prosandcons",
    "sectors",
    "financial_ratios",
    "peer_groups",
    "stock_prices"
]

for table in tables:
    count = pd.read_sql(
        f"SELECT COUNT(*) AS c FROM {table}",
        conn
    ).iloc[0,0]

    summary["Table"].append(table)
    summary["Rows"].append(count)

pd.DataFrame(summary)
```

# Cell 15 – Close Connection

```python
conn.close()

print("Database Closed")
```
